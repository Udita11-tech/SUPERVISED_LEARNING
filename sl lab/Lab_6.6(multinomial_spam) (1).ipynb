```python
# prepare a multinomial naive-bayes' classification model for test classification of email messages.
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score, f1_score
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Latin-1 encodeing is used for assigning a unique numerical value to each character which includes many non-ASCII characters as well
df = pd.read_csv('spam.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df = df[['v1', 'v2']]
df.head()
```

```python
df = df.rename(columns={'v1': 'label', 'v2': 'text'})
df.head()
```

```python
x=df['text']
y=df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Value_counts is used to count the numnber of unique values in a dataset
distribution = y.value_counts()
distribution
```

```python
distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title("Distrbution of Spam and Ham messages")
plt.show()
```

```python
# Generate Word cloud for Spam Messages
spam_text = ' '.join(df[df['label'] == 'spam']['text'])
spam_text
```

```python
# Generate the Wordcloud for the spam emails
# A wordcloud in python is a visual representation of text data that uses the size and color of words to show their frequency
spam_wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white', random_state=42).generate(spam_text)
```

```python
# Generate Word cloud for Ham Messages
ham_text = ' '.join(df[df['label'] == 'ham']['text'])
ham_text
```

```python
# Generate the Wordcloud for the Ham emails
ham_wordcloud = WordCloud(width=800, height=400, max_words=100, background_color='white', random_state=42).generate(ham_text)
```

```python
# Plot the wordcloud for spam meassages
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(spam_wordcloud)
plt.title('Word Cloud for Spam Messages')
plt.axis('off')

# Plot the wordcloud for ham messages
plt.subplot(1, 2, 2)
plt.imshow(ham_wordcloud)
plt.title('Word Cloud for Ham Messages')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Create a CountVectorizer to convert text data into numerical features
# Count vectorizer is a text processing technique which is used in nlp task for converting a collection of text into numerical represnetaion
vectorizer = CountVectorizer()

x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
```

```python
# Train a Multinomial Naive Bayes classifier
model_multinomial = MultinomialNB(alpha = 0.8, fit_prior = True, force_alpha = True)
model_multinomial.fit(x_train, y_train)
```

```python
# Train a Gaussian Naive Bayes classifier
model_gaussian = GaussianNB()
model_gaussian.fit(x_train.toarray(), y_train)
```

```python
# Calculating the Accuracy
y_pred_multinomial = model_multinomial.predict(x_test)
accuracy_multinomial = accuracy_score(y_test, y_pred_multinomial)
print("Accuracy for Multinomial Naive Bayes Model: ", accuracy_multinomial)

y_pred_gaussian = model_gaussian.predict(x_test.toarray())
accuracy_gaussian = accuracy_score(y_test, y_pred_gaussian)
print("Accuracy for Gaussian Naive Bayes Model: ", accuracy_gaussian)
```

```python
methods = ["Multinomial Naive Bayes", "Gaussian Naive Bayes"]
scores = [accuracy_multinomial, accuracy_gaussian]
plt.bar(methods, scores)
plt.xlabel("Classification Methods")
plt.ylabel("Accuracy")
plt.title("Comparison of Classification Methods")
plt.show()
```
