```python
#wap in python to print the series of prime numbers from 1 to 300 and print in series
l=[]
for i in range(1,301):
    if i>1:
        for j in range(2,i//2+1):
            if i%j==0:
                break
        else:
            l.append(i)
primes=pd.Series(l)
primes
```

```python
#fibonacci
a,b=0,1
l=[]
while a<100:
    l.append(a)
    a,b=b,a+b
fibonaccis=Series(l)
fibonaccis
```

```python
#wap to add all the numbers in a user defined list of 20 elements and find min and max amongst that list without using functions
l=[int(x) for x in input("Enter 20 numbers: ").split()]
min=l[0]
sum=0
max=l[0]

for i in l:
    sum+=i
    if i<min:
        min=i
    if i>max:
        max=i

print("Sum: ",sum)
print("Min: ",min)
print("Max: ",max)
```
